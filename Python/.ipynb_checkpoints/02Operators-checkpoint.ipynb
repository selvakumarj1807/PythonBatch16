# Operators

Operations on some of Values or Variables.

# Types of Operators

1. Arithmatic Operators
2. Arithmatic Assignment Operators
3. Relational or Comparision Operator
4. Logical Operators
5. Membership Operators
6. Identity Operators
7. Bitwise Operators

# 1. Arithmatic Operators

```python
a = 7

b = 2

print('Addition (+): ', a + b)

print('Subtraction (-): ', a - b)

print('Multiplication (*): ', a * b)

print('Divsion (/): ', a / b) # Quotient Value

print('-----------------------------')

print('Floor Divsion (//): ', a // b)

print('Modulas (%): ', a % b) # Remainder Value

print('Exponentiation (**): ', a ** b) # 7 ** 2 = 7 * 7 = 49
```

# 2 Arithmatic Assignment Operators

```python
# Assignment Operators

a = 10

print('Assignment Operators(=): ', a)

# Arithmatic Assignment Operators

a += 10 # a = a + 10

print('Arithmatic Assignment Operators (+=): ', a)
```

# 3. Relational or Comparision Operator

```python
a = 10

b = 5

print('Equal (==): ', a == b)

print('Not Equal (!=): ', a != b)

print(a > 10)

print(a >= 10)

print(a < 10)

print(a <= 10)
```

# 4. Logical Operators

```python
userName = input('Enter your UserName : ')

password = input('Enter your Password : ')

print(userName == 'admin')

print(password == 'admin123')

print('------------------------------')

# Logical And Operators
# Any one Condition have False. Then Everything is False

print((userName == 'admin') and (password == 'admin123'))

print('------------------------------')

# Logical Or Operators
# Any one Condition have True. Then Everything is True

print((userName == 'admin') or (password == 'admin123'))

print('------------------------------')

# Logical Not Operators
# True => False, False => True

print (not(userName == 'admin'))

print (not(password == 'admin123'))
```
