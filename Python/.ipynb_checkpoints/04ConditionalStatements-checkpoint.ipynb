# If

```python
age = 20

if age >= 18:
    print(age, ' Above 18. So you are eligible for vote')
```

# If & Else

```python
age = 12

if age >= 18:
    print(age, ' Above 18. So you are eligible for vote')

else:
    print(age, ' Below 18. So you are not eligible for vote')
```

# If & Elif & Else

```python
n = 0

if n > 0:
    print('Positive Value')
    
elif n < 0:
    print('Negotive Value')
    
else:
    print('Zero')
```

# Using Logical Operators
