# Example Program

The area of a circle is the total space inside its boundary, calculated using the formula A = \(A = \pi r^2\)

A  = πr^2

```python
pi = 3.14

r = 105

A = pi * r * r

print('The area of a circle radius of', r, 'is' , A)

```
