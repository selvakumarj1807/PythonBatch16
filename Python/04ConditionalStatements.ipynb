# If

```python
age = 20

if age >= 18:
    print(age, ' Above 18. So you are eligible for vote')
```

# If & Else

```python
age = 12

if age >= 18:
    print(age, ' Above 18. So you are eligible for vote')

else:
    print(age, ' Below 18. So you are not eligible for vote')
```

# If & Elif & Else

```python
n = 0

if n > 0:
    print('Positive Value')
    
elif n < 0:
    print('Negotive Value')
    
else:
    print('Zero')
```

# Nested IF

```python
mark01 = 21
mark02 = 23
mark03 = 22

if mark01 >= 35 and mark02 >= 35 and mark03 >= 35:
    total = mark01 + mark02 + mark03
    average = int(total / 3)

    print('Total : ', total)

    print('Average : ', average)

    if average >= 90 and average <= 100:
        print('A Grade')

    elif average >= 80 and average <= 90:
        print('B Grade')

    elif average >= 70 and average <= 80:
        print('C Grade')
        
    else:
        print('No Grade')

else:
    print('You have no Grade & Total & Average')

    if mark01 <= 35:
        print(mark01, ' - Mark 01 is Below 35')

    if mark02 <= 35:
        print(mark02, ' - Mark 02 is Below 35')

    if mark03 <= 35:
        print(mark03, ' - Mark 03 is Below 35')
```
